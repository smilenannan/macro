```python
# model(3)
import pandas as pd
import statsmodels.api as sm
import numpy as np

data_path = "data/caschool.xls"
data = pd.read_excel(data_path, sheetname='caschool')

X = data[['str', 'el_pct']]
y = data['testscr']

HiEL = X['el_pct'].apply(lambda x: 1 if x>=10 else 0)
HiEL.name = 'HiEL'
HiEL_by_str = X['str'] * HiEL
HiEL_by_str.name = 'HiEL_by_str'

X = pd.concat([X, HiEL, HiEL_by_str], axis=1)[['str', 'HiEL', 'HiEL_by_str']]
mod = sm.OLS(y, sm.add_constant(X))
res = mod.fit()

rob_res = res.get_robustcov_results()
print(rob_res.summary())

print('SER : ', np.sqrt(rob_res.scale))

A = np.array([[0, 1, 0, 0], [0, 0, 0, 1]])
print(rob_res.f_test(A))
```

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
from dateutil.relativedelta import relativedelta
import numpy as np

f = '%Y:%m'
my_parser = lambda date: pd.datetime.strptime(date, f)

data_path = "data/oj.xls"
data = pd.read_excel(data_path, sheetname='Sheet2', index_col=0, parse_dates=0, date_parser=my_parser)
price = 100 * data['PPIOJ']/data['PWFSA']

all_pct_delta = []
start_date = dt.datetime.strptime('1950-01-01', '%Y-%m-%d')
end_date = dt.datetime.strptime('2000-12-01', '%Y-%m-%d')

current = start_date
while current < end_date:
    former = current
    current += relativedelta(months=1)
    pct_delta = 100 * (np.log(price.loc[current])- np.log(price.loc[former]))
    all_pct_delta.append(pct_delta)

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
plt.plot(price.index.tolist()[1:], all_pct_delta)
ax.set_xticks([dt.datetime.strptime('1950-01-01', '%Y-%m-%d'), dt.datetime.strptime('2000-12-01', '%Y-%m-%d')])
ax.set_xticklabels(['1950-01', '2000'])
ax.set_xlabel('時間')
ax.set_ylabel('オレンジジュース実質価格の%変化[%]')
plt.savefig("2-1-b.png", format='png', dpi=300)
```

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm

f = '%Y%m'
my_parser = lambda date: pd.datetime.strptime(date, f)

adj_data_path = "data/2015taxadj.xls"
adj_data = pd.read_excel(adj_data_path, sheetname='zmy', index_col=0, parse_dates=0, date_parser=my_parser, skiprows=[1, 2, 3, 4, 5])
adj_criinf = adj_data['食料（酒類を除く）及びエネルギーを除く総合']
adj_criinf = adj_criinf.iloc[12:]

noadj_data_path = "data/zmy2015a.csv"
noadj_data = pd.read_csv(noadj_data_path, index_col=0, encoding="shift-jis", parse_dates=0, date_parser=my_parser, skiprows=[1, 2, 3, 4, 5])
noadj_criinf = noadj_data['食料（酒類を除く）及びエネルギーを除く総合']
noadj_before_criinf = noadj_criinf.loc['1983-01-01':'1995-12-01']
noadj_after_criinf = noadj_criinf.loc['2016-01-01':'2016-12-01']

criinf = pd.concat([noadj_before_criinf, adj_criinf, noadj_after_criinf])

quarter_criinf = []
i = 0
while i<len(criinf):
    quarter_criinf.append((criinf.iloc[i]+criinf.iloc[i+1]+criinf.iloc[i+2])/3)
    i += 3
    
quarter_criinf = pd.Series(quarter_criinf)
    
ygap_data_path = "data/gap.xlsx"
ygap_data = pd.read_excel(ygap_data_path, sheetname='Chart', index_col=0)
ygap = ygap_data.ix[3:-2, '需給ギャップ']

quarter_criinf.index = ygap.index

# test
# prepare data
test_data = pd.DataFrame()
test_data['ygap'] = ygap.astype(float)
test_data['quarter_criinf'] = quarter_criinf
test_data['flag'] = (test_data.index < "1996.1Q")
# test two regression lines
from statsmodels.formula.api import ols

formula = 'quarter_criinf ~ ygap * C(flag)' 
lm = ols(formula, test_data)
res = lm.fit()
rob_res = res.get_robustcov_results()
print (rob_res.summary())

# test for slope and intercept
A = np.array([[0, 1, 0, 0], [0, 0, 0, 1]])
print(rob_res.f_test(A))
```
